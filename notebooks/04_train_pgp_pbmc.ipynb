#### Using the model

```python
# --------------------------------------------------
# Load + align Leiden labels
# --------------------------------------------------
leiden_path = data_path / "rna_leiden_labels.csv"
leiden_df = pd.read_csv(leiden_path)
leiden_df.head()
```

```python
from multiomic_vae.training.pgp_train import pgp_train

# IMPORTANT: PGP is Peak -> Gene -> Peak, so training forward uses peak as input.
model, gene_x, peak_x = pgp_train(gene, peak)
```

```python
%run ../multiomic_vae/training/pgp_train.py
```

```python
from multiomic_vae.evaluation.pgp_test import run_tests_with_wandb
```

```python
from sklearn.preprocessing import LabelEncoder

# If cell IDs are a column, move them to index
if "cell_id" in leiden_df.columns:
    leiden_df = leiden_df.set_index("cell_id")

# Align to gene dataframe order (gene & peak should have the same index ordering)
leiden_df_aligned = leiden_df.loc[gene.index]

# Encode cluster labels as integers
labels = LabelEncoder().fit_transform(
    leiden_df_aligned.iloc[:, 0].astype(str).values
)
```

```python
print((leiden_df_aligned.index == gene.index).all())
```

```python
# --------------------------------------------------
# Run evaluation + log to wandb
# --------------------------------------------------
results = run_tests_with_wandb(
    model=model,
    gene_tensor=gene_x,
    peak_tensor=peak_x,
    labels=labels
)

results
```
